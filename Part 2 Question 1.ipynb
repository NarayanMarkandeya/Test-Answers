```python
import language_tool_python
tool = language_tool_python.LanguageTool('en-US')
```

```python
a=[]
for i in df1.index:
    text = df1['text'][i]
    matches = tool.check(text)
    count = len(matches)
    a.append(count)
    if count == 0:
        print(text," -- No mistakes")
    else:
        print(text," -- Mistakes found, ",count," mistakes")

tool.close()
```

```python
print(a)
```

```python
df1['No_of_Mistakes'] = a
```

Added 1 column No_of_Mistakes at last in df1 dataframe. Which shows the number of gramatical mistakes, will be easy to understand.
